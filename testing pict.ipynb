```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
import numpy as np
import os
import cv2
```

```python
DataDirectory = r"C:\Users\Windows 11\Documents\Train Emotion\archive\train"
```

```python
Classes = ["angry", "disgust", "fear", "happy", "neutral", "sad", "surprise"] #ListClass
```

```python
num_images_to_show = 12

fig, axes = plt.subplots(nrows=1, ncols=num_images_to_show, figsize=(10, 5))

# Iterasi untuk setiap kategori
for category in Classes:
    path = os.path.join(DataDirectory, category)
    img_files = os.listdir(path)
    
    for i, img in enumerate(img_files[:num_images_to_show]):
        img_array = cv2.imread(os.path.join(path, img))
        
        # Jika ada gambar untuk ditampilkan
        if img_array is not None:
            # Mengubah warna dari BGR ke RGB
            img_rgb = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
            
            # Menampilkan gambar pada subplot
            ax = axes[i]  # Mengakses subplot ke-i
            ax.imshow(img_rgb)
            ax.axis('off') 
            ax.set_title(f'{category} - {i+1}')
    
    plt.tight_layout()
    plt.show()
    break  
```
